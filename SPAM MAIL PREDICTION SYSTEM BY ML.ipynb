```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the data from csv file to a pandas Dataframe
raw_mail_data = pd.read_csv("mail_data.csv")
```

```python
print(raw_mail_data)
```

```python
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# printing the first 5 rows of the dataframe
mail_data.head()
```

```python
mail_data.tail()
```

```python
# checking the number of rows and columns in the dataframe
mail_data.shape
```

```python
mail_data.info()
```

```python
mail_data.isnull().sum()
```

```python
# label spam mail as 0;  ham mail as 1;

mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
# separating the data as texts and label

X = mail_data['Message']

Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# prediction on test data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
input_mail = ["free"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```

```python
import tkinter as tk
from tkinter import messagebox, scrolledtext, filedialog
from PIL import Image, ImageTk
import pickle
from sklearn.feature_extraction.text import CountVectorizer
import pyttsx3
from datetime import datetime
```

```python
# === Load Model and Vectorizer ===
with open('spam.pkl', 'rb') as f:
    model1 = pickle.load(f)

with open('vectorizer.pkl', 'rb') as f:
    cv = pickle.load(f)
```

```python
# === Speak Function ===
def speak(message):
    engine = pyttsx3.init()
    engine.say(message)
    engine.runAndWait()
```

```python
# === Dummy User Database ===
USERS = {
    "ASHU": "1234",
    "emma": "abcd"
}
```

```python
# === Main App Window ===
root = tk.Tk()
root.title("SPAM MAIL PREDICTION")
root.state('zoomed')  # Full screen
```

```python
# === Background Image ===
try:
    bg_img = Image.open("NIGHT.jpg")
    bg_img = bg_img.resize((root.winfo_screenwidth(), root.winfo_screenheight()))
    bg_photo = ImageTk.PhotoImage(bg_img)
    bg_label = tk.Label(root, image=bg_photo)
    bg_label.place(relwidth=1, relheight=1)
except Exception as e:
    print("Error loading background image:", e)
```

```python
# === App Frames ===
login_frame = tk.Frame(root, bg="#ffffff", bd=0)
app_frame = tk.Frame(root, bg="#ffffff", bd=2)
for frame in (login_frame, app_frame):
    frame.place(relwidth=1, relheight=1)
```

```python
# === Centered Login UI Container ===
login_container = tk.Frame(login_frame, bg="white")
login_container.place(relx=0.5, rely=0.5, anchor="center")

tk.Label(login_container, text="ðŸ” SPAM PREDICTION", font=("Helvetica", 28, "bold"),
         bg="white", fg="#8a2be2").pack(pady=30)

tk.Label(login_container, text="USER NAME", bg="white", font=("Helvetica", 16, "bold")).pack()
username_entry = tk.Entry(login_container, font=("Helvetica", 14), width=25)
username_entry.pack(pady=10)

tk.Label(login_container, text="PASSWORD", bg="white", font=("Helvetica", 16, "bold")).pack()
password_entry = tk.Entry(login_container, font=("Helvetica", 14), show="*", width=25)
password_entry.pack(pady=10)
```

```python
def login():
    user = username_entry.get().strip()
    pw = password_entry.get().strip()
    if user in USERS and USERS[user] == pw:
        current_user["name"] = user
        welcome_label.config(text=f"Welcome, {user.capitalize()} ðŸ‘‹")
        speak(f"Hello {user}, welcome to SPAM PREDICTION SYSTEM.")
        messagebox.showinfo("Welcome", f"Hello {user}, welcome to the Spam Prediction!")
        animated_transition(app_frame)
    else:
        messagebox.showerror("Login Failed", "Invalid username or password")
        speak("Login failed. Try again.")
```

```python
def register():
    user = username_entry.get().strip()
    pw = password_entry.get().strip()
    if not user or not pw:
        messagebox.showwarning("Missing Info", "Username and password required.")
        return
    if user in USERS:
        messagebox.showwarning("Exists", "User already exists.")
        return
    USERS[user] = pw
    messagebox.showinfo("Success", f"User {user} registered successfully.")
    speak("Registered successfully.")

btn_frame = tk.Frame(login_container, bg="white")
btn_frame.pack(pady=20)

tk.Button(btn_frame, text="Login", command=login,
          font=("Helvetica", 16, "bold"), bg="#6a5acd", fg="white",
          padx=20, pady=8, bd=0).grid(row=0, column=0, padx=10)

tk.Button(btn_frame, text="Register", command=register,
          font=("Helvetica", 16, "bold"), bg="#20b2aa", fg="white",
          padx=20, pady=8, bd=0).grid(row=0, column=1, padx=10)
```

```python
# === Frame Switching with Animation ===
def animated_transition(target_frame):
    root.update()
    login_frame.lower()
    target_frame.tkraise()
```

```python
# === User Data ===
current_user = {"name": ""}
history = []
stats = {"total": 0, "spam": 0, "ham": 0}
```

```python
# === App Layout ===
tk.Label(app_frame, text="SPAM MAIL PREDICTION SYSTEM", font=("Helvetica", 32, "bold"),
         bg="white", fg="#ff4500").pack(pady=20)

welcome_label = tk.Label(app_frame, text="", font=("Helvetica", 20, "bold"),
                         bg="white", fg="#006400")
welcome_label.pack()

tk.Label(app_frame, text="ENTER YOUR MAIL", bg="white",
         font=("Helvetica", 16, "bold")).pack(pady=(20, 5))

text_input = tk.Entry(app_frame, font=("Helvetica", 14,"bold"), width=80, relief="solid", bd=2)
text_input.pack(pady=10, ipady=15)

button_frame = tk.Frame(app_frame, bg="white")
button_frame.pack(pady=10)

tk.Button(button_frame, text="Classify", command=lambda: classify(),
          font=("Helvetica", 16, "bold"), bg="#32cd32", fg="white",
          padx=25, pady=10, bd=0).pack(side="left", padx=(0, 15))

tk.Button(button_frame, text="ðŸ“‚ Open File", command=lambda: load_file(),
          font=("Helvetica", 14, "bold"), bg="#ffa500", fg="black",
          padx=20, pady=10, bd=0).pack(side="left", padx=(15, 0))

result_label = tk.Label(app_frame, text="", font=("Helvetica", 16, "bold"),
                        bg="white", fg="black")
result_label.pack(pady=10)

icon_label_left = tk.Label(app_frame, bg="white")
icon_label_left.place(x=30, y=30)

icon_label_right = tk.Label(app_frame, bg="white")
icon_label_right.place(x=root.winfo_screenwidth() - 170, y=30)
```

```python
# === Stats Section ===
stats_section = tk.Frame(app_frame, bg="#e6f0fa", bd=1, relief="solid")
stats_section.pack(pady=10, padx=60, fill="x")

pad_val = 120
total_label = tk.Label(stats_section, text="Total: 0", font=("Helvetica", 15, "bold"), bg="#e6f0fa")
spam_label = tk.Label(stats_section, text="Spam: 0", font=("Helvetica", 15, "bold"), bg="#e6f0fa", fg="red")
ham_label = tk.Label(stats_section, text="Not Spam: 0", font=("Helvetica", 15, "bold"), bg="#e6f0fa", fg="green")
percentage_label = tk.Label(stats_section, text="Spam %: 0%", font=("Helvetica", 15, "bold"), bg="#e6f0fa")

total_label.pack(side="left", padx=pad_val)
spam_label.pack(side="left", padx=pad_val)
ham_label.pack(side="left", padx=pad_val)
percentage_label.pack(side="left", padx=pad_val)
```

```python
# === History ===
tk.Label(app_frame, text="ðŸ•“ Classification History:", font=("Helvetica", 16, "bold"),
         bg="white", fg="#444").pack(pady=(25, 5))

history_box = scrolledtext.ScrolledText(app_frame, height=8, width=100, font=("Consolas", 11,"bold"), bg="#f9f9f9")
history_box.pack(pady=5)
```

```python
# === Update Stats ===
def update_stats():
    total = stats["total"]
    spam = stats["spam"]
    ham = stats["ham"]
    percentage = (spam / total * 100) if total > 0 else 0

    total_label.config(text=f"Total: {total}")
    spam_label.config(text=f"Spam: {spam}")
    ham_label.config(text=f"Not Spam: {ham}")
    percentage_label.config(text=f"Spam %: {percentage:.1f}%")
```

```python
# === Classify Function ===
def classify():
    msg = text_input.get().strip()
    if not msg:
        messagebox.showwarning("Empty Input", "Please enter a message to classify.")
        return

    vect = cv.transform([msg]).toarray()
    prediction = model1.predict(vect)[0]
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    stats["total"] += 1

    icon_label_left.config(image='')
    icon_label_right.config(image='')

    if prediction == 1:
        message = "ðŸš« This is a SPAM mail!"
        color = "red"
        stats["spam"] += 1
        try:
            spam_img = Image.open("SPAM IMAGE.jpg").resize((190, 190))
            spam_icon = ImageTk.PhotoImage(spam_img)
            icon_label_left.config(image=spam_icon)
            icon_label_left.image = spam_icon
        except Exception as e:
            print("Error loading SPAM IMAGE:", e)
    else:
        message = "âœ… This is NOT a spam mail."
        color = "green"
        stats["ham"] += 1
        try:
            ok_img = Image.open("OK IMAGE.jpg").resize((190, 190))
            ok_icon = ImageTk.PhotoImage(ok_img)
            icon_label_right.config(image=ok_icon)
            icon_label_right.image = ok_icon
        except Exception as e:
            print("Error loading OK IMAGE:", e)

    result_label.config(text=message, fg=color)
    speak(message)

    log = f"[{now}] {'SPAM' if prediction == 1 else 'NOT SPAM'} >> {msg}\n"
    history.append(log)
    history_box.insert(tk.END, log)
    history_box.see(tk.END)

    update_stats()
```

```python
# === Load File Function ===
def load_file():
    file_path = "mail_data.csv"
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()
    except Exception as e:
        messagebox.showerror("Error", f"Failed to read file: {e}")
        return

    file_window = tk.Toplevel(root)
    file_window.title("Select Text to Classify")
    file_window.state('zoomed')

    content_box = scrolledtext.ScrolledText(file_window, wrap="word", font=("Consolas", 13))
    content_box.pack(expand=True, fill="both", padx=20, pady=20)
    content_box.insert(tk.END, content)

    def use_selected_text():
        try:
            selected_text = content_box.selection_get().strip()
            if selected_text:
                text_input.delete(0, tk.END)
                text_input.insert(0, selected_text)
                file_window.destroy()
            else:
                messagebox.showwarning("No Selection", "Please select some text from the file.")
        except tk.TclError:
            messagebox.showwarning("No Selection", "Please select some text from the file.")

    tk.Button(file_window, text="Use Selected Text", command=use_selected_text,
              font=("Helvetica", 14, "bold"), bg="#1e90ff", fg="white",
              padx=20, pady=10).pack(pady=15)
```

```python
# === Start the App ===
login_frame.tkraise()
root.mainloop()
```
